```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import KMeans
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv")
```

```python

df = pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
x = df.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters = i, init="k-means++", random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("no.of.clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters=5, init = 'k-means++', random_state = 0)
y_kmeans = kmeansmodel.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans == 0,0],x[y_kmeans == 0,1], s=100, c='red', label="centroid 1")
plt.scatter(x[y_kmeans == 1,0],x[y_kmeans == 1,1], s=100, c='blue', label="centroid 2")
plt.scatter(x[y_kmeans == 2,0],x[y_kmeans == 2,1], s=100, c='green', label="centroid 3")
plt.scatter(x[y_kmeans == 3,0],x[y_kmeans == 3,1], s=100, c='grey', label="centroid 4")
plt.scatter(x[y_kmeans == 4,0],x[y_kmeans == 4,1], s=100, c='yellow', label="centroid 5")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s = 300, c="pink", label="centroids")
plt.title("Clusters of customers")
plt.xlabel("Annual Income(k$)")
plt.ylabel("Spending score(1-100)")
plt.legend()
plt.show()
```
